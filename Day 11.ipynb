# List Comprehensions

```python
a = 1,2,3,4,5
a
```

```python
[i for i in a]
```

```python
[i**2 if i%2==0 else i**3 for i in range(1,11)] #list comprehension return square if i is odd else cube
```

```python
{i**2 if i%2==0 else i**3 for i in range(1,11)}
```

```python
[i**2 if i%2==0 else i**3 for i in range(1,11) if i%5!=0]
```

Dict Comprehension

```python
{i:i**2 if i%2==0 else i**3 for i in range(1,11)}
```

```python
{i:i**2 if i%2==0 else i**3 for i in range(1,11) if i%5!=0}
```

```python
[i for i in range(1,20) if i%2==0]
```
