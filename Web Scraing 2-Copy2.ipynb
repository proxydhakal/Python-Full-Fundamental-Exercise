```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request
base_url='https://old.reddit.com/top/'
request = urllib.request.Request(base_url)
html = urllib.request.urlopen(request).read()
soup = BeautifulSoup(html,'lxml')
divs =soup.findAll('div')
contentDiv =soup.find('div',attrs={'id':'siteTable'})
lastpage=soup.findAll('a',attrs={'rel':'nofollow next'})
len(lastpage)
rank = [data.text for data in contentDiv.findAll('span',attrs={'class':'rank'})]
title =[data.text for data in contentDiv.findAll('a',attrs={'data-event-action':'title'})]
comments = [data.text for data in contentDiv.findAll('a',attrs={'data-event-action':'comments'})]
username = [data.text for data in contentDiv.findAll('a',attrs={'class':'author'})]
test_df = pd.DataFrame({'Rank': rank,
    'Title': title,
    'Comments': comments,
    'Username': username
    })
test_df   
export_csv = test_df.to_csv (r'export_data.csv', index = None, header=True)    

```

```python
csvFields =["Rank","Title","Total Comments","Username"]
import csv
with open('extractData.csv', 'w',newline='') as wf:
    csvWriter =csv.writer(wf)
    csvWriter.writerow(csvFields)
    csvWriter.writerow([rank,title,comments,username])
print("\nData Written to CSV File Sucessfully")
    
```
