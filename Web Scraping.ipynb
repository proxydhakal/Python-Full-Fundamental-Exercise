# Web Scraping

```python
base_url='http://stats.espncricinfo.com/ci/engine/records/batting/most_runs_career.html?id=12357;type=tournament'
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
responce =requests.get(base_url)
```

```python
responce.status_code
```

```python
html=responce.content
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
tables = soup.findAll('table')
len(tables)
```

```python
table = soup.find('table', attrs={"class":"engineTable"})
```

```python
all_rows= table.findAll('tr')
```

```python
all_rows
```

```python
len(all_rows)
```

```python
raw_columns, *raw_rows = all_rows
```

```python
raw_rows
```

```python
columns =[data.text for data in raw_columns.findAll('th')]
```

```python
columns
```

```python
rows =[[data.text for data in row.findAll('td')]for row in raw_rows]
```

```python
rows
```

```python
datas =[dict(zip(columns,row)) for row in rows]
```

```python
datas
```

```python
import csv
with open('players_data.csv','w', newline='') as wf:
    writer = csv.DictWriter(wf, fieldnames=columns)
    writer.writeheader()
    writer.writerows(datas)
```

2. Python program to count the number of string and integer in a string:

```python
string = input("Enter your string:")
intCount=0
strCount = 0
for i in string:
    if i.isnumeric():
        intCount+=1
    else:
        strCount+=1
print("The total number of Integer and String in a string are",intCount,',',strCount,'.')
```

3. Python program to check if a string is a Pangram or Not.

```python
from string import ascii_lowercase as a_l
```

```python
def check(s):
    return set(a_l)-set(s.lower())==set([])
string = input("Enter String:")
if check(string)==True:
    print("The string is Pangram.")
else:
    print("The string is not Pangram.")
```
